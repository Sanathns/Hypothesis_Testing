```python
unitA=pd.Series(data.iloc[:,0])
unitA
```

```python
unitB=pd.Series(data.iloc[:,1])
unitB
```

```python
# 2-sample 2-tail ttest:  stats.ttest_ind(array1,array2)
pvalue=stats.ttest_ind(unitA,unitB)
pvalue
```

#Ans: Take Null hyposthesis as Ho: μ1 = μ2 (There is no difference in diameters of cutlets between two units) and take Alternate hypothesis as Ha: μ1 ≠ μ2 (There is significant difference in diameters of cutlets between two units) 2 Sample 2 Tail test applicable here pvalue is 0.4722 which is greater than alpha(0.5) so we failed reject null hypothesis

```python
pvalue[1]
```

```python
if pvalue[1]>0.05:
  print("Accept null Hypothesis")
else:
  print("Accept Alternate hypothesis")
```
